Importo las librerias que crea necesarias


```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Accidentes = pd.read_csv("../Datasets/LimpiezaFinal.csv")
AccidentesFinal = Accidentes.drop(["Unnamed: 0"], axis="columns")
```

Numero total de fallecidos

```python
AccidentesFatales = AccidentesFinal[AccidentesFinal.Fatalities > 0]
AccidentesFatales.shape
total_fatalities = AccidentesFinal.Fatalities.sum()
total_fatalities
```

```python
AccidentesFinal['Date'] = AccidentesFinal['Date'].apply(pd.to_datetime)
datetimes = AccidentesFinal.Date
datetimes
```

```python
AccidentesFinal['year'] = AccidentesFinal['Date'].dt.year
```

```python
AccidentesFinal['year'].value_counts(ascending = True)
```

```python
crash_by_year = AccidentesFinal['year'].value_counts(ascending = True) \
                                 .plot(kind='bar', figsize=(15,7))

for bars in crash_by_year.containers:
    crash_by_year.bar_label(bars, label_type='edge')

plt.title('Crashes by Year', fontsize=30)
plt.ylabel('Number of Crashes', fontsize=20)
plt.xlabel('Year', fontsize=20);

plt.savefig("CrashesByYear.jpg")
```
