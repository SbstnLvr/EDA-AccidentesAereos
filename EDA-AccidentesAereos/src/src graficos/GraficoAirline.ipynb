```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Accidentes = pd.read_csv("LimpiezaFinal.csv")
```

```python
AccidentesFinal = Accidentes.drop(["Unnamed: 0"], axis="columns")
```

```python
AccidentesFinal
```

```python
AccidentesFatales = AccidentesFinal[AccidentesFinal.Fatalities > 0]
AccidentesFatales
```

```python
AccidentesDomesticos = AccidentesFinal[AccidentesFinal['Aircraft Type'].str.contains("Passenger|Domestic",na=True)]
AccidentesDomesticos
```

```python
AccidentesDomesticos_by_airline = AccidentesDomesticos['Airline'].value_counts(ascending = False) \
                                  .head(15) \
                                  .sort_values(ascending=True) \
                                  .plot(kind='barh', figsize=(15,7))
                                
plt.title('Crashes by Airline',fontsize=30)
plt.ylabel('Airlines', fontsize=20)
plt.xlabel('Number of Crashes', fontsize=20);

plt.savefig("CrashesByAirline.jpg")
```

```python
AccidentesDomesticosFatales = AccidentesDomesticos[AccidentesDomesticos.Fatalities > 0]
AccidentesDomesticosFatales
```

```python
AccidentesDomesticosFatales_by_airline = AccidentesDomesticosFatales['Airline'].value_counts(ascending = False) \
                                  .head(15) \
                                  .sort_values(ascending=True) \
                                  .plot(kind='barh', figsize=(15,7))
                                
plt.title('Fatal Crashes by Airline',fontsize=30)
plt.ylabel('Airlines', fontsize=20)
plt.xlabel('Number of Fatal Crashes', fontsize=20);

plt.savefig("FatalCrashesByAirline.jpg")
```
