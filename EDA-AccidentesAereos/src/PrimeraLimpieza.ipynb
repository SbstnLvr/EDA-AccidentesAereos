Importo las librerias que crea necesarias para la limpieza de datos 

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

Leemos el csv con el que trabajaremos

```python
Accidentes = pd.read_csv("../Datasets/Aircraft_Incident_Dataset.csv")
```

En este caso ya venian mis datos casi limpios pero retirare ciertos datos que no necesitaremos en esta ocasión y cambiaré de nombre a algunas columnas por comodidad o faltas de ortografía

```python
Accidentes.columns
```

```python
Accidentes.drop(['Aircaft_Registration',"Date", "Onboard_Crew", "Onboard_Passengers", "Aircaft_First_Flight","Time","Arit","Aircaft_Engines","Ground_Casualties","Collision_Casualties"], inplace=True, axis=1)
Accidentes
```

```python
Accidentes = Accidentes.rename(columns={"Incident_Date":"Date","Aircaft_Model":"Aircraft Model","Aircaft_Operator":"Airline",
                           "Aircaft_Nature":"Aircraft Type","Incident_Category":"Incident Type",
                            "Incident_Cause(es)":"Incident Cause","Incident_Location":"Incident Location",
                            "Aircaft_Damage_Type":"Damage","Onboard_Total":"Onboard","Aircraft_Phase":"Phase",
                            "Departure_Airport":"Departure","Destination_Airport":"Destination" })
Accidentes
```

Voy a reducir manualmente los datos para que sea solo los ultimos 15 años, (teniendo en cuenta que la fecha de los datos que tenemos es hasta 2022)

```python
Accidentes = Accidentes[0:3043]
Accidentes
```

Cambiamos valores vacios como ?

```python
Accidentes = Accidentes.replace({"?":"Unknown","Info-Unavailable":"Unavailable"})
Accidentes
```

Ordenamos de fecha 2007 a 2022

```python
Accidentes = Accidentes[::-1]
Accidentes
```

Guardamos esta primera limpieza

```python
Accidentes.to_csv("PrimeraLimpieza.csv")
```
