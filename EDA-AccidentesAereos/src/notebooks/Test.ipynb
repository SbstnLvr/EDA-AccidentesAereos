Importo las librerias que crea necesarias para la limpieza de datos 

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Accidentes = pd.read_csv("../Datasets/LimpiezaFinal.csv")
Accidentes
```

```python
AccidentesFinal = Accidentes.drop(["Unnamed: 0"], axis="columns")
```

```python
AccidentesFinal
```

```python
AccidentesFatales = AccidentesFinal[AccidentesFinal.Fatalities > 0]
AccidentesFatales.shape
total_fatalities = AccidentesFinal.Fatalities.sum()
total_fatalities
```

```python
AccidentesFinal['Date'] = AccidentesFinal['Date'].apply(pd.to_datetime)
datetimes = AccidentesFinal.Date
datetimes
```

```python
AccidentesFinal['month'] = AccidentesFinal['Date'].dt.month_name()
AccidentesFinal['Day of Week'] = AccidentesFinal['Date'].dt.day_name()
AccidentesFinal['year'] = AccidentesFinal['Date'].dt.year
```

```python
AccidentesFinal['month'].value_counts(ascending = True)
```

```python
crash_by_day = AccidentesFinal['month'].value_counts(ascending = True) \
                                 .plot(kind='bar', figsize=(15,7))

for bars in crash_by_day.containers:
    crash_by_day.bar_label(bars, label_type='edge')

plt.title('Crashes by Month', fontsize=30)
plt.ylabel('Number of Crashes', fontsize=20)
plt.xlabel('Name of the Month', fontsize=20);
```

```python
AccidentesFinal['year'].value_counts(ascending = False).sort_values()
```

```python
crash_by_day = AccidentesFinal['year'].value_counts(ascending = True) \
                                 .plot(kind='bar', figsize=(15,7))

for bars in crash_by_day.containers:
    crash_by_day.bar_label(bars, label_type='edge')

plt.title('Crashes by Year', fontsize=30)
plt.ylabel('Number of Crashes', fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
AccidentesFinal['Day of Week'].value_counts(ascending = True)
```

```python
crash_by_day = AccidentesFinal['Day of Week'].value_counts(ascending = True) \
                                 .plot(kind='bar', figsize=(15,7))

for bars in crash_by_day.containers:
    crash_by_day.bar_label(bars, label_type='edge')

plt.title('Crashes by Day of Week', fontsize=30)
plt.ylabel('Number of Crashes', fontsize=20)
plt.xlabel('Days of the Week', fontsize=20);
```
